```python
def train_test_split_scale(X, y, category_features, test_size=0.2, scale: bool=True):
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import StandardScaler
    # Split the whole dataset to train and validate
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0, shuffle=False) 
    num_feaures =  [x for x in X_train.columns if x not in category_features]

    if scale:
        scaler = StandardScaler()
        
        # transform train and test sets
        scaler.fit(X_train[num_feaures])
        X_train_scaled = scaler.transform(X_train[num_feaures])
        X_train_num = pd.DataFrame(X_train_scaled, index=X_train.index, columns=num_feaures)
#         X_train = pd.concat([X_train_num, X_train[category_features]], axis=1)
        
        scaler.fit(X_test[num_feaures])
        X_test_scaled = scaler.transform(X_test[num_feaures])
        X_test_num = pd.DataFrame(X_test_scaled, index=X_test.index, columns=num_feaures)
#         X_test = pd.concat([X_test_num, X_test[category_features]], axis=1)
        
    return X_train, X_test, y_train, y_test
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import BaggingClassifier

def select_features_seq(X, y, direction='forward', info_sets=None):
    base_clf = RandomForestClassifier(
        criterion='entropy',
        n_estimators=256,
        max_leaf_nodes=16,
        max_features=1,
        bootstrap=False,
        class_weight='balanced_subsample',
        n_jobs=-1,
        verbose=False)

    seq_fs = SequentialFeatureSelector(base_clf, n_features_to_select=8,
                                    n_jobs=n_cpus, direction=direction,
                                    cv=None, scoring='neg_log_loss')   
    seq_fs.fit(X, y)
    return seq_fs

seq_fs = select_features_seq(X_train,
                         y_train,
                         direction='forward',
                         info_sets=None)

print(X_train.columns[seq_fs.get_support()])
```

## Calculate Feature Importances

```python
def calc_one_feature_importance(clf, X, y, triple_barrier_events, method='MDA', n_splits=10):
    oob_score = clf.oob_score_
    sample_events = triple_barrier_events.dropna().loc[X.index,'t1']
    
    cv_gen = PurgedKFold(n_splits=n_splits,
                         samples_info_sets=sample_events,
                         pct_embargo=0.01)
    
    oos_score = ml_cross_val_score(clf, X, y, cv_gen=cv_gen,
                                   scoring=accuracy_score,
                                   require_proba=False).mean()

    if method == 'MDI':
        imp = mean_decrease_impurity(clf, X.columns)
    elif method == 'MDA':
        imp = mean_decrease_accuracy(clf, X, y, cv_gen, scoring=accuracy_score, require_proba=False)
    elif method == 'SFI':
        imp = single_feature_importance(clf, X, y, cv_gen, scoring=accuracy_score, require_proba=False)
    
    return imp, oob_score, oos_score


def calc_all_feature_importances(clf, X, y, triple_barrier_events, n_splits=10):
    oob_score = clf.oob_score_
    sample_events = triple_barrier_events.dropna().loc[X.index,'t1']
    
    cv_gen = PurgedKFold(n_splits=n_splits,
                         samples_info_sets=sample_events,
                         pct_embargo=0.01)
    
    oos_score = ml_cross_val_score(clf, X, y, cv_gen=cv_gen,
                                   scoring=accuracy_score,
                                   require_proba=False).mean()

    mdi_imp = mean_decrease_impurity(clf, X.columns)
    mda_imp = mean_decrease_accuracy(clf, X, y, cv_gen, scoring=accuracy_score, require_proba=False)
    sfi_imp = single_feature_importance(clf, X, y, cv_gen, scoring=accuracy_score, require_proba=False)
    
    return mdi_imp, mda_imp, sfi_imp, oob_score, oos_score
```

```python
def get_uncorrelated_features(X_train, y_train, correlated_groups):
    from sklearn.ensemble import RandomForestClassifier
    uncorrelated_feat = pd.DataFrame()
    for group in correlated_groups:
    #     group = correlated_groups[9]
        # add all features of the group to a list
        feature_list = list(group['feature2'].unique())+ list(group['feature1'].unique())
        feature_list = list(set(feature_list))
        # train a random forest 
        rf = RandomForestClassifier(n_estimators=200, random_state=39, max_depth=4)
        rf.fit(X_train[feature_list], y_train)

        # get the feature importance attributed by the 
        # random forest model (more on this in coming lectures)

        importance = pd.concat(
            [pd.Series(feature_list),
             pd.Series(rf.feature_importances_)], axis=1)

        importance.columns = ['feature', 'importance']

        # sort features by importance, most important first
        importance = importance.sort_values(by='importance', ascending=False)
        uncorrelated_feat = pd.concat([uncorrelated_feat, importance.iloc[0,:]], axis=1)
    return uncorrelated_feat
```
